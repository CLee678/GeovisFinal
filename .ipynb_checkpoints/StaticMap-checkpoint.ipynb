```python
# needed imports
import geopandas as gpd
import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib import colors
from matplotlib import cm
from mpl_toolkits.axes_grid1 import make_axes_locatable
```

```python
# load the file in GeoDataFrame
population_file = r"data/Vital_Population_CT.geojson"
population = gpd.read_file(population_file)
population.head()
```

```python
# get columns name
population.columns
```

```python
# describe the data
population.describe
```

```python
# get the Coordinate Reference System(CRS) of the data
population.crs
```

```python
# if it doesn't changes to 2272, the shape looks distorted, so it's better changes to Philadelphia's EPSG 2272.
population_2272 = population.to_crs(epsg=2272) 
population_2272.crs
```

```python
# prepare the axis and coordinate
fig, ax = plt.subplots(2, 3, figsize=(12, 11))

# tittle
fig.suptitle('Race & Ethnicity in Philadelphia', fontsize=24, fontname='Times New Roman', fontweight='bold')

# plot White
population_2272.plot(column='PERCENT_WHITE_NH', cmap='RdPu', edgecolor='0.5', ax=ax[0][0], linewidth=0.5, k=10, scheme='equal_interval')
ax[0][0].set_title('White', fontsize='16', fontname='Times New Roman')
ax[0][0].axis('off')

# plot Black
population_2272.plot(column='PERCENT_BLACK_NH', cmap='RdPu', edgecolor='0.5', ax=ax[0][1], linewidth=0.5, k=10, scheme='equal_interval')
ax[0][1].set_title('Black', fontsize='16', fontname='Times New Roman')
ax[0][1].axis('off')

# plot Asian
population_2272.plot(column='PERCENT_ASIAN_PI_NH', cmap='RdPu', edgecolor='0.5', ax=ax[0][2], linewidth=0.5, k=10, scheme='equal_interval')
ax[0][2].set_title('Asian', fontsize='16', fontname='Times New Roman')
ax[0][2].axis('off')  # Remove axis labels and ticks

# plot Hispanic
population_2272.plot(column='PERCENT_HISPANIC', cmap='RdPu', edgecolor='0.5', ax=ax[1][0], linewidth=0.5, k=10, scheme='equal_interval')
ax[1][0].set_title('Hispanic', fontsize='16', fontname='Times New Roman')
ax[1][0].axis('off')

# plot Multiracial
population_2272.plot(column='PERCENT_MULTIRACIAL_NH', cmap='RdPu', edgecolor='0.5', ax=ax[1][1], linewidth=0.5, k=10, scheme='equal_interval')
ax[1][1].set_title('Multiracial', fontsize='16', fontname='Times New Roman')
ax[1][1].axis('off')

# remove the last dataframe
ax[1][2].axis('off')

# create colorbar
cbar_ax = fig.add_axes([0.9, 0.3, 0.02, 0.4])

# set the color and normalization
cmap = plt.cm.RdPu
norm = colors.Normalize(vmin=0, vmax=100)

# create a ScalarMappable to set the colorbar
sm = plt.cm.ScalarMappable(cmap=cmap, norm=norm)
sm.set_array([])

# add colorbar with custom ticks and labels
cbar = plt.colorbar(sm, cax=cbar_ax, ticks=[0, 25, 50, 75, 100])
cbar.set_label('Percentage', rotation=270, labelpad=10)

# adjust layout
plt.tight_layout(rect=[0, 0.05, 0.9, 0.95])

# save the plot
fig.savefig('Race & Ethnicity in Philadelphia.png', dpi=300)

# show the plot
plt.show()
```
